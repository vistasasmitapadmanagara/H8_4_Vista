## Cleaning Data

Sebelum analisa kita lakukan, mari kita bersihkan terlebih dahulu data kita, agar hasil analisa yang lebih baik dan tidak terganggu missing values maupun invalid values.

Mari kita drop terlebih dahulu kolom EASE-MEAN, Unnamed: 0,  
- EASE-MEAN tidak memiliki data.
- Unnamed: 0 berisikan data yang tidak diperlukan.

```python
df.drop(columns=['EASE-MENT', 'Unnamed: 0'], axis=1, inplace=True)
```

```python
df.isna().sum()
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['APARTMENT NUMBER'].fillna(method='bfill', inplace=True)

df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(keep=False, inplace=True)

df.duplicated().sum()
```

```python
df.rename(columns={'BOROUGH': 'DISTRIK',
                   'NEIGHBORHOOD':'DAERAH_SEKITAR',
                   'BUILDING CLASS CATEGORY': 'KATEGORI_KELAS_BANGUNAN',
                   'TAX CLASS AT PRESENT': 'KELAS_PAJAK_SAAT_INI',
                   'BLOCK': 'BLOK', 
                   'LOT': 'KAVLING',
                   'BUILDING CLASS AT PRESENT': 'KELAS_BANGUNAN_SAAT_INI',
                   'ADDRESS': 'ALAMAT',
                   'APARTMENT NUMBER': 'NOMOR_APARTEMEN',
                   'ZIP CODE': 'KODE_POS',
                   'RESIDENTIAL UNITS': 'UNIT_PERUMAHAN',
                   'COMMERCIAL UNITS': 'UNIT_KOMERSIAL',
                   'TOTAL UNITS': 'JUMLAH_UNIT',
                   'LAND SQUARE FEET': 'UKURAN_TANAH_PERSEGI',
                   'GROSS SQUARE FEET': 'UKURAN_RUANGAN_PERSEGI',
                   'YEAR BUILT': 'TAHUN_DIBANGUN',
                   'TAX CLASS AT TIME OF SALE': 'KELAS_PAJAK_PADA_SAAT_PENJUALAN',
                   'BUILDING CLASS AT TIME OF SALE': 'KELAS_BANGUNAN_PADA_SAAT_PENJUALAN',
                   'SALE PRICE': 'HARGA_PENJUALAN',
                   'SALE DATE': 'TANGGAL_PENJUALAN'}, inplace=True)

df['HARGA_PENJUALAN'] = df['HARGA_PENJUALAN'].astype(int)

df.head()
```
