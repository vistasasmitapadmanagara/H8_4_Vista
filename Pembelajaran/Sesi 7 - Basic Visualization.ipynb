Kelihatannya tidak benar!

Situasi seperti ini sering terjadi saat membuat plot, solusinya sering kali terletak pada bagaimana kumpulan data yang mendasarinya disusun.

Mari kita rubah urutan dataset, dan kemudian memplot kembali.

```python
# transpose dataframe
df_trans = df.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_trans.head()
```

```python
df_trans.plot(kind='hist', figsize=(10,6))

plt.title('Histogram of Immigration from Denmark, Norway, Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

Buat beberapa modifikasi untuk meningkatkan impact dan estetika dari plot sebelumnya.

```python
count, bin_edges = np.histogram(df_trans, 15)

df_trans.plot(kind='hist',
              figsize=(10,6),
              bins=15,
              alpha=0.6,
              xticks=bin_edges,
              color=['coral', 'darkslateblue', 'mediumseagreen'])

plt.title('Histogram of Immigration from Denmark, Norway, Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
import matplotlib
for name, hex in matplotlib.colors.cnames.items():
    print(name, hex)
```

Jika tidak ingin plot saling tumpang tindih, kita dapat menumpuknya menggunakan parameter <b>stacked</b>. Mari kita juga menyesuaikan label sumbu x min dan maks untuk menghilangkan celah ekstra di tepi plot. Kita bisa pass tuple(min, max) menggunakan parameter <b>xlim</b>.

```python
count, bin_edges = np.histogram(df_trans, 15)
xmin = bin_edges[0] - 10
xmax = bin_edges[-1] + 10

df_trans.plot(kind='hist',
              figsize=(10,6),
              bins=15,
              xticks=bin_edges,
              color=['coral', 'darkslateblue', 'mediumseagreen'],
              stacked=True,
              xlim=(xmin, xmax))

plt.title('Histogram of Immigration from Denmark, Norway, Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

## Pie Charts

<b>Question:</b> Gunakan pie chart untuk menjelajahi proporsi (persentase) imigran baru yang dikelompokkan berdasarkan benua untuk seluruh periode waktu dari 1980 hingga 2013.

```python
# group countries by continents and apply sum() function
df_continents = df.groupby('Continent', axis=0).sum()

print(type(df.groupby('Continent', axis=0)))

df_continents.head()
```

```python
df_continents['Total'].plot(kind='pie',
                            figsize=(5,6),
                            autopct='%1.1f%%',
                            startangle=90,
                            shadow=True)

plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal')

plt.show()
```

Visual di atas tidak begitu jelas, angka dan teks tumpang tindih dalam beberapa kasus. Mari buat beberapa modifikasi untuk meningkatkan visual:

```python
color_list=['gold',
             'yellowgreen',
             'lightcoral',
             'lightskyblue',
             'lightgreen',
             'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1]

df_continents['Total'].plot(kind='pie',
                            figsize=(15, 8),
                            autopct='%1.1f%%',
                            startangle=90,
                            shadow=True,
                            labels=None,
                            pctdistance=1.12,
                            colors=color_list,
                            explode=explode_list)

plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.12)
plt.axis('equal')

plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

<b>Question:</b> Using a pie chart, explore the proportion (percentage) of new immigrants grouped by continents in the year 2013.

```python
color_list=['gold',
             'yellowgreen',
             'lightcoral',
             'lightskyblue',
             'lightgreen',
             'pink']
explode_list = [0.1, 0, 0, 0, 0.05, 0.15]

df_continents['2013'].plot(kind='pie',
                            figsize=(15, 8),
                            autopct='%1.1f%%',
                            startangle=90,
                            shadow=True,
                            labels=None,
                            pctdistance=1.12,
                            colors=color_list,
                            explode=explode_list)

plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.12)
plt.axis('equal')

plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

## Bubble Plots (Optional)

<b>Question:</b> Let's analyses the effect of this crisis, and compare Argentina's immigration to that of it's neighbour.

```python
df_trans = df[years].transpose()
df_trans.index = map(int, df_trans.index)
df_trans.index.name = 'Year'
df_trans.reset_index(inplace=True)

df_trans.head()
```

```python
# normalize Brazil data
norm_brazil = (df_trans['Brazil'] - df_trans['Brazil'].min()) / (df_trans['Brazil'].max() - df_trans['Brazil'].min())

# normalize Argentina data
norm_argentina = (df_trans['Argentina'] - df_trans['Argentina'].min()) / (df_trans['Argentina'].max() - df_trans['Argentina'].min())
```

```python
ax0 = df_trans.plot(kind='scatter',
                    x='Year',
                    y='Brazil',
                    figsize=(14, 8),
                    alpha=0.5,
                    color='green',
                    s=norm_brazil * 2000 + 10,
                    xlim = (1975, 2015))

ax1 = df_trans.plot(kind='scatter',
                    x='Year',
                    y='Argentina',
                    figsize=(14, 8),
                    alpha=0.5,
                    color='blue',
                    s=norm_argentina * 2000 + 10,
                    ax=ax0)

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 - 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```
